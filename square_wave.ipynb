```python
## import packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
## f(x) exact
def f_even_square(x, L): 
    fxn = np.piecewise(x, [(-L <= x) & (x < -L/2), 
                        (-L/2 <= x) & (x < L/2), 
                        (L/2 <= x) & (x <= L)], 
                       [-1, 1, -1]) 
    return fxn

## Fourier cosine series approx to f(x)
def fourier_cos(x, N, L):
    a0 = 0
    an = lambda n: 4/(n*np.pi)*(-1)**((n-1)/2) if n%2 != 0 else 0
    cosn = lambda n: np.cos(n*np.pi*x/L)
    f = a0
    for n in range(1,N):
        f = f + an(n)*cosn(n)
    
    return f

## set parameters
L = 2

## create x-grid
xmin = -L
xmax = +L
nsteps = 100001
x = np.linspace(xmin, xmax, nsteps)

## plot
plt.plot(x, f_even_square(x, L), color='grey', label='exact')
plt.plot(x, fourier_cos(x, 4, L), 'b--', label='first 4 terms')
plt.plot(x, fourier_cos(x, 10, L), 'm-.', label='first 10 terms')
plt.plot(x, fourier_cos(x, 100, L), 'r:', label='first 100 terms')
plt.legend(loc='lower right')
plt.show()
```
