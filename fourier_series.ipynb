```python
## import packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
## f(x)=|x| exact
def f_even(x):
    return abs(x)

## Fourier cosine series approx to f(x)
def fourier_cos(x, N):
    a0 = np.pi/2
    an = lambda n: -4/(n**2 * np.pi) if n%2 != 0 else 0
    cosn = lambda n: np.cos(n*x)
    f = a0
    for n in range(1,N):
        f = f + an(n)*cosn(n)
    
    return f

## create x-grid
xmin = -np.pi
xmax = +np.pi
nsteps = 101
x = np.linspace(xmin, xmax, nsteps)

## plot
plt.plot(x, f_even(x), color='grey', label='exact')
plt.plot(x, fourier_cos(x, 2), 'b--', label='first 2 terms')
plt.plot(x, fourier_cos(x, 4), 'm-.', label='first 4 terms')
plt.plot(x, fourier_cos(x, 8), 'r:', label='first 8 terms')
plt.legend(loc='lower right')
plt.show()
```

```python
## f(x)=x exact
def f_odd(x):
    return x

## Fourier sine series approx to f(x)
def fourier_sin(x, N):
    bn = lambda n: -2/n*(-1)**(n)
    sinn = lambda n: np.sin(n*x)
    f = 0
    for n in range(1,N):
        f = f + bn(n)*sinn(n)
    
    return f

## create x-grid
xmin = -np.pi
xmax = +np.pi
nsteps = 101
x = np.linspace(xmin, xmax, nsteps)

## plot
plt.plot(x, f_odd(x), color='grey', label='exact')
plt.plot(x, fourier_sin(x, 2), 'r--', label='first 2 terms')
plt.plot(x, fourier_sin(x, 4), 'm-.', label='first 4 terms')
plt.plot(x, fourier_sin(x, 8), 'b:', label='first 8 terms')
plt.legend(loc='lower right')
plt.show()
```
