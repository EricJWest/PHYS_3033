```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def f_even(x):
    return abs(x)

def fourier_cos(x, N):
    a0 = np.pi/2
    an = lambda n: -4/(n**2 * np.pi)
    cosn = lambda n: np.cos(n*x)
    f = a0
    for n in range(1,N):
        f = f + an(n)*cosn(n)
    
    return f

xmin = -np.pi
xmax = +np.pi
nsteps = 101
x = np.linspace(xmin, xmax, nsteps)

plt.plot(x, fourier_cos(x, 2), 'r--', label='first 2 terms')
plt.plot(x, fourier_cos(x, 4), 'm-.', label='first 4 terms')
plt.plot(x, fourier_cos(x, 500), 'b:', label='first 500 terms')
plt.plot(x, f_even(x), 'k-', label='exact')
plt.legend(loc='lower right')
plt.show()
```

```python
def f_odd(x):
    return x

def fourier_sin(x, N):
    bn = lambda n: -2/n*(-1)**(n)
    sinn = lambda n: np.sin(n*x)
    f = 0
    for n in range(1,N):
        f = f + bn(n)*sinn(n)
    
    return f

xmin = -np.pi
xmax = +np.pi
nsteps = 101
x = np.linspace(xmin, xmax, nsteps)

plt.plot(x, fourier_sin(x, 2), 'r--', label='first 2 terms')
plt.plot(x, fourier_sin(x, 4), 'm-.', label='first 4 terms')
plt.plot(x, fourier_sin(x, 500), 'b:', label='first 500 terms')
plt.plot(x, f_odd(x), 'k-', label='exact')
plt.legend(loc='lower right')
plt.show()
```
