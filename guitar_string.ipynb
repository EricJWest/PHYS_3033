```python
## import packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
## f(x) exact
def f(x, L, A):
    slope = 2*A/L
    intercpt = 2*A
    fxn = np.piecewise(x, [x < L/2, x >= L/2], 
                       [lambda x: slope*x, lambda x: -slope*x + intercpt]) 
    return fxn

## Fourier sine series approx to f(x)
def fourier_sine(x, L, A, N):
    bn = lambda n: 8*A/(n*np.pi)**2 * (-1)**((n-1)/2) if n%2 !=0 else 0
    sinn = lambda n: np.sin(n*np.pi*x/L)
    f = 0
    for n in range(1,N):
        f = f + bn(n)*sinn(n)
    
    return f

## set physical parameters
A = 3
L = 65

## create x-grid
xmin = 0
xmax = L
nsteps = 101
x = np.linspace(xmin, xmax, nsteps)

## plot
plt.plot(x, f(x, L, A), 'k-', label='exact')
plt.plot(x, fourier_sine(x, L, A, 2), 'r--', label='first 2 terms')
plt.plot(x, fourier_sine(x, L, A, 4), 'm-.', label='first 4 terms')
plt.plot(x, fourier_sine(x, L, A, 10), 'b:', label='first 10 terms')
plt.legend(loc='upper right')
plt.show()
```
